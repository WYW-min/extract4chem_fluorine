```python
INPATH = "/Data_two/wyw/data/JTSJ/待修复"
OUTPATH = "/Data_two/wyw/data/JTSJ/待修复_展开"
```

```python
from tqdm.notebook import tqdm
from pathlib import Path
recorded_ids = set()
to_read_paths = []
for p in Path(INPATH).glob("**/*.md"):
    if p.stem in recorded_ids:
        continue
    recorded_ids.add(p.stem)
    to_read_paths.append(p)
```

```python
for p in tqdm(to_read_paths):
    outpath = Path(OUTPATH) / p.name
    outpath.write_bytes(p.read_bytes())
```
