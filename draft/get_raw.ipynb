```python
from pathlib import Path
import jsonlines
from datetime import datetime
now_date_str = datetime.now().strftime("%Y%m%d%H%M%S")
```

```python
import hashlib
def str_to_md5(input_str: str, encoding: str = "utf-8") -> str:
    """
    将字符串转换为MD5哈希值（32位十六进制字符串）
    
    Args:
        input_str: 需要转换的原始字符串
        encoding: 字符串编码格式，默认utf-8
    
    Returns:
        32位小写的MD5十六进制字符串
    """
    try:
        str_bytes = input_str.encode(encoding)
        md5_obj = hashlib.md5()
        md5_obj.update(str_bytes)
        md5_hex = md5_obj.hexdigest()
        return md5_hex
    except UnicodeEncodeError as e:
        raise ValueError(f"字符串编码失败，编码格式{encoding}不支持：{e}")
```

```python
indir = Path("../data/raw/test-聚酰亚胺文献/md_enhance")
outdir = Path("../data/out/聚酰亚胺")
outpath = outdir / f"raw_{now_date_str}.jsonl"
with jsonlines.open(outpath, "w") as writer:
    for inpath in indir.glob("./*.md"):
        writer.write({
            "file_name" : inpath.name,
            "md5" : str_to_md5(inpath.read_text()),
            "content" : inpath.read_text()
        })
    
```
